```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head(50)
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
selected_features = df.drop("koi_disposition", axis=1)
selected_features
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
y = df["koi_disposition"]
X = selected_features
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42)
X_train
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Import dependencies
from sklearn.preprocessing import LabelEncoder, MinMaxScaler,StandardScaler
from tensorflow.keras.utils import to_categorical

# Scale your data
from sklearn.preprocessing import StandardScaler
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test) 

# Label-encode data
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
print(y_test)
print(encoded_y_test)
```

# Train the Model



```python
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression(max_iter=10000)
model1.fit(X_train_scaled, encoded_y_train)

predictions = model1.predict(X_test_scaled)
print(predictions)

print(f"Training Data Score: {model1.score(X_train_scaled, encoded_y_train)}")
print(f"Testing Data Score: {model1.score(X_test_scaled, encoded_y_test)}")
```

```python
model1
```

```python
from sklearn.metrics import classification_report
print(classification_report(encoded_y_test, predictions, 
                            target_names = ["Confirmed", "False Positive", "Candidate"]))
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [10, 50, 100],
             'max_iter':[200, 500, 1000]}
grid = GridSearchCV(model1, param_grid, verbose=3)
```

```python
# Train the model with GridSearch
grid.fit(X_train_scaled, encoded_y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
